### 11.2.1 - Visualizing two quantitative variables

In the last plot, we looked at the relationship between the total bills and tips at a restaurant. We highlighted two subgroups. male and female, with the *hue* parameter, by using a different colors on the same plot. 

There is another method to highlight subgroups, by creating separate plots per each subgroup. To do this, we need the *seaborn* function *relplot*, which stands for "relational plots". We can use *relplot* to display two quantitative variable by usign either scatter plots or line plots. 

**Scatter plots**

One of the best features of *relplot* is the possibility to create subplots in a single figure. This is an example of how to use *relplot*, by using the same data *tips*:

```python
sns.relplot(x = "total_bill",       # change the function name to relplot
               y= "tip", 
               data=tips,
               kind="scatter",      # kind parameter is used to specify which type of plots you want 
               col="sex")           # obtain separate scatter plots for male and female, arranged horizontally in columns
```

```python
sns.relplot(x = "total_bill",     
               y= "tip", 
               data=tips,
               kind="scatter",      
               row="sex")           # scatter plots for male and female arranged vertically in rows
```

```python
sns.relplot(x = "total_bill",     
               y= "tip", 
               data=tips,
               kind="scatter",
               col= "time",         # col= & row= can be used at the same time
               row="sex")           # to obtain a subplot for each combination of time&sex 
```

```python
sns.relplot(x = "total_bill",     
               y= "tip", 
               data=tips,
               kind="scatter",
               col= "day",         
               col_wrap=2,                  # specifies how many subplots you want per row
               col_order=["Thur", 
                           "Fri",
                           "Sat",
                           "Sun"])           # to change the order of the subplots by giving a list of ordered values
```

## 12.1 - Two samples t-test

One of the most widely known basic bivariate statistical tests is the independent t-test (aka student's t-test), which tests whether there is a statistically significant difference between the means in two unrelated groups. 

We'll use the Iris dataset for this example, which includes information on the sepal and petal length and width of different flower species. 

```python
import pandas as pd
df = pd.read_csv("C:\\Users\\lucp8040\\Documents\\Flames\\Python intro flames\\datasets\\Iris_Data.csv") 
df.head() #check the first rows of your data
```

It's always a good idea to have a first look at your data using some descriptives. The *.describe()* function is one way of doing this.

```python
#some descriptive statistics
df.groupby("species")["sepal_width"].describe()
```

Let's focus on Iris-setosa and Iris-versicolor. There seems to be a difference between the average sepal width and the standard deviations are small. But is this difference significant?

Before we move to a statistical test, we need to check some assumptions. 


### 12.1.1 - Assumption checks
#### Homogeneity of variances

```python
# Create 2 DataFrames as subsets (they only contains data for flower species)
setosa = df[(df['species'] == 'Iris-setosa')] #dataframe only containing setosa flowers
setosa.reset_index(inplace= True) #reset index of new dataframe!
versicolor = df[(df['species'] == 'Iris-versicolor')] #dataframe only containing versicolor flowers
versicolor.reset_index(inplace= True) #reset index of new dataframe
```

Levene's test for homogeneity of variance is widely used. 

```python
from scipy import stats #import the stats functions from the scipy package

stats.levene(setosa['sepal_width'], versicolor['sepal_width'])
```

The p-value is 0.42, which means that according to Levene's test there is no reason to suspect a difference in variances between the two groups. We can proceed.

#### Normally distributed residuals

Checking for normality can be done visually (e.g. p-p plot, q-q plot, or histogram) or using a statistical test (e.g. Shapiro-Wilk test to formally test for normality).

Let's first check with a P-P plot. 

```python
# Normally distributed residuals
#calculate residuals
diff = setosa['sepal_width'] - versicolor['sepal_width'] 
import matplotlib.pyplot as plt
stats.probplot(diff, plot= plt)
plt.title('Sepal Width P-P Plot') 
```

Another option is a simple histogram of the residuals

```python
plt.hist(diff)
```

Or, alternatively, a Shapiro-Wilk test

```python
stats.shapiro(diff)
```

The first value is the W test statistic, and the second one the p-value. Given this p-value, we are confident to conclude that the distribution of the residuals does not differ significantly from a normal distribution.

### 12.1.2 - Two samples t-test

```python
stats.ttest_ind(setosa['sepal_width'], versicolor['sepal_width'])
```

Conclusion: 

Iris-setosas average sepal width (M= 3.418, SD= 0.381) is wider and has slightly larger variation than Iris-versicolor (M= 2.770, SD= 0.314). An independent t-test was used to test for a difference. Results indicate that we are confident to conclude that there is a statistically significant difference between the sepal width of Iris-setosa and Iris-versicolor (t(98)= 9.282, p< 0.0001).

## 12.2 - Chi square test

Another well known test is the chi-square test, which is used to test the independence of two categorical variables. 

We'll be looking into the relationship between 'sex' and 'hours-per-week'. We assume the data contains a random sample from the population and is a good representation of the population. Furthermore, each persons can be in one category for each variable only (i.e. each individual can only have one 'sex').

```python
import pandas as pd
workhour_by_sex = pd.read_csv('C:\\Users\\lucp8040\\Documents\\Flames\\Python intro flames\\datasets\\workinghours.csv')
workhour_by_sex = workhour_by_sex[['sex', 'hours_per_week_categories']]
workhour_by_sex.head()
```

Let's check the division of workhours and of workhours by sex.

```python
workhour_by_sex['hours_per_week_categories'].value_counts()
```

```python
workhour_by_sex['sex'].value_counts()
```

### 12.2.1 - Constructing the contingency table

A first step is to construct the frequency count table, aka the contingency table. This is the crosstable with the observed frequencies, used to calculate the chi square value.

```python
contingency_table = pd.crosstab(
    workhour_by_sex['sex'],
    workhour_by_sex['hours_per_week_categories'],
    margins = True)
contingency_table
```

We can visualize it with a stacked barplot.

```python
#Assigns the frequency values
malecount = contingency_table.iloc[0][0:6].values
femalecount = contingency_table.iloc[1][0:6].values


import matplotlib.pyplot as plt
import seaborn as sns  

#Plots the bar chart
fig = plt.figure(figsize=(10, 5))
sns.set(font_scale=1.8)
categories = ["0-9","10-19","20-29","30-39","40-49","50+"]
p1 = plt.bar(categories, malecount, 0.55, color='#d62728')
p2 = plt.bar(categories, femalecount, 0.55, bottom=malecount)
plt.legend((p2[0], p1[0]), ('Male', 'Female'))
plt.xlabel('Hours per Week Worked')
plt.ylabel('Count')
plt.show()
```

### 12.2.2 - Calculating the Chi square test


Using *scipy*, we can now calculate the chi-squared test on this contingency table. 
All you need to do is convert the observed frequencies into a two-dimensional **array**, which is necessary for the scipy function.

```python
import numpy as np
f_obs = np.array([contingency_table.iloc[0][0:6].values,
                  contingency_table.iloc[1][0:6].values])
f_obs
```

```python
from scipy import stats
stats.chi2_contingency(f_obs)[0:3]
```

The results show that chi square =~ 2287, the associated p-value =~ 0, and df=5. 

Conclusion:
The p-value is <0.05, so we are confident to reject the null hypothesis. There is evidence of a relationship between 'sex' and 'hours-per-week'. However, we don't know the direction of this relationship, or which categories are associated with each other. We only now that these two variables are not independent of each other.

## 12.4 - Simple linear regression

Regression searches for relationships among variables. The term "linear" refers to a linear relationship between variables. If we draw this relationship in a two-dimensional space, we get a straight line.

There are several possible Python packages that you can use to run a linear regression. Here, we will use again the package *scipy*. To estimate the slope of the line, we are going to use *linregress* from the *scipy.stats* module:

```python
from scipy.stats import linregress
```

```python
# Get the data "tips"
import seaborn as sns
tips=sns.load_dataset('tips')
tips.head()
```

There are two types of variables: one dependent (response) and one or more independent (predictors) variables. Linear regression performs the task to predict a dependent variable value (y) based on a given independent variable (x). In our example, we would like to set up a regression model to predict *tip* by *total_bill*.

```python
# Extract the variables
subset = tips.dropna(subset=['total_bill', 'tip'])  # Linregress cannot handle missing values so drop them
xs = subset['total_bill']
ys = subset['tip']
```

```python
# Run regression with scipy
res = linregress(xs, ys)
res
```

The result is a *linregress* result object, that contains five values: 
    
       - slope: is the slope of the line of best fit for our data; 
       - intercept: is the intercept;
       - rvalue: is the correlation coefficient;
       - pvalue: p-values of the slope; 
       - stderr: standard error of the slope.
       
The fitted model for the tips at the restaurant as a function of total bill is:

tip = 0.92 + 0.10* total_bill

How do we interpret the slope coefficient? Each additional 1$ increment in the total bill is associated with about a 0.10 (10 cents) increase in the tip. 

To better understand the slope, we can plot the line of best fit by using the *seaborn* method *lmplot*: 

```python
sns.lmplot(x ='total_bill', y ='tip', data = tips)
```

Another option to run linear regressions is the package *statsmodels*. Its a powerful Python package for the estimation of statistical models, performing tests, and more. Indeed, in case you want to run a multiple regression, you have to use *statsmodels* since *scipy* doesn't do multiple regression. You can find more information on *statsmodels* on its official web site: https://www.statsmodels.org/stable/index.html.

Let's run the same regression using *statsmodels*, and to confirm the same results. We compute the regression of *tip* as a function of *total_bill* using *statsmodels*' *smf.ols()*:

```python
import statsmodels.api as sm
X=tips['total_bill'] #define independent var
Y=tips['tip'] #define dependent var

X = sm.add_constant(X) #adds a constant for the model
X.head()
```

```python

model = sm.OLS(Y,X).fit()
predictions=model.predict(X)

print_model = model.summary()
print(print_model)
```

```python
dir(model) #gives you all the info present in the fitted model object, you can access them all for exploration
```

```python
print('Adj R squared =', model.rsquared_adj) #for adjusted r squared
```

```python
print('Predicted values =', model.predict()) #for predicted values
```

Lets test for normality of residuals using omnibus test:

```python
import statsmodels.stats.api as sms   #the .stats version of statsmodel gives some extra diagnostics options
test = sms.omni_normtest(model.resid)
test
```

Furthermore, you can run a linear regression in Python relatively easily by using the package *scikit-learn* as well. It is a widely used Python library for machine learning, built on top of *NumPy*. It provides the means for preprocessing data, reducing dimensionality, implementing regression, classification, clustering, and more. The procedure is similar to that of *statsmodels*.
